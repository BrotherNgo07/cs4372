```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import tensorflow as tf
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/a-nagar/datasets/main/pima_indian_diabetes.csv", header = 0)
```

```python
df
```

```python
X = df.drop("Outcome", axis = 1)
```

```python
X
```

```python
y = df['Outcome']
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
```

```python
y = to_categorical(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
model = Sequential()
model.add(Dense(500, activation='relu', input_dim=8))
model.add(Dense(100, activation='relu'))
model.add(Dense(2, activation='softmax'))

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train)
```

```python
model.summary()
```

```python

scores = model.evaluate(X_train, y_train, verbose=0)
print('Accuracy on training data: {}% \n Error on training data: {}'.format(scores[1], 1 - scores[1]))
y_test_pred = model.predict(X_test)
scores2 = model.evaluate(X_test, y_test, verbose=0)
print('Accuracy on test data: {}% \n Error on test data: {}'.format(scores2[1], 1 - scores2[1]))
```

## Asian Cuisines Dataset

```python
cuisines_df = pd.read_csv("https://an-utd-python.s3.us-west-1.amazonaws.com/cuisines.csv")
cuisines_df.head()
```

```python
cuisines_label_df = cuisines_df['cuisine']
cuisines_label_df.head()
```

```python
cuisines_feature_df = cuisines_df.drop(['Unnamed: 0', 'cuisine'], axis=1)
cuisines_feature_df.head()
```

```python
X = cuisines_feature_df
y = cuisines_label_df
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y = encoder.fit_transform(y)
y = to_categorical(y)
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
model = Sequential()
model.add(Dense(500, activation='relu', input_dim=380))
model.add(Dense(200, activation='relu'))
model.add(Dense(100, activation='relu'))

model.add(Dense(5, activation='softmax'))

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train)
```

```python

scores = model.evaluate(X_train, y_train, verbose=0)
print('Accuracy on training data: {}% \n Error on training data: {}'.format(scores[1], 1 - scores[1]))
y_test_pred = model.predict(X_test)
scores2 = model.evaluate(X_test, y_test, verbose=0)
print('Accuracy on test data: {}% \n Error on test data: {}'.format(scores2[1], 1 - scores2[1]))
```
